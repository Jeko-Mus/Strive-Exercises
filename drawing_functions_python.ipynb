```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
```

1. Plot the absolute value of x and its derivative. Choose an interval to have a nice looking shape.

```python
x = np.linspace(-100,100,100)
y = 1 / (np.abs(x))
#plt.plot(x,y)

def absol(x):
    return 1 / (np.abs(x))

abs_derivative = derivative(absol, x, dx=1e-6)
plt.plot(x, abs_derivative)
```

2. Plot the function $x\log(x)$ and its derivative.

```python
x = np.linspace(1,100,100)
y = np.log(x)
#plt.plot(x,y)

def log(x):
    return np.log(x)

log_derivative = derivative(log, x, dx=1e-6)
plt.plot(x, log_derivative)
```

3. Plot the function $e^{x}\log(x)+5$ and its derivative.

```python
x = np.linspace(1,100,1000)
y = np.exp(x) * np.log(x) + 5 
#plt.plot(x,y)

def logexp(x):
    return np.exp(x) * np.log(x) + 5 

logexp_derivative = derivative(logexp, x, dx=1e-6)
plt.plot(x, logexp_derivative)
```
